```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flights = pd.read_csv('/flights_data.csv')
print(flights.shape)
flights.head()
```

**Bar Charts:**

```python
plt.figure(figsize=(14,6))
sns.countplot(data=flights, x='Source')
plt.ylabel('Number of Flights', fontsize=12)
plt.xlabel('Source', fontsize=12)
```

```python
plt.figure(figsize=(14,6))
base_color = sns.color_palette()[0]
sns.countplot(data = flights, x = 'Source', color = base_color)
plt.xticks(rotation=30)
```

```python
plt.figure(figsize=(14,6))
base_color = sns.color_palette()[1]
gen_order = flights['Source'].value_counts().index
sns.countplot(data = flights, x = 'Source', color = base_color, order = gen_order)
```

```python
plt.figure(figsize=(14,6))
base_color = sns.color_palette()[2]
sns.countplot(data = flights, x = 'Airline', color = base_color)
```

```python
plt.figure(figsize=(14,6))
base_color = sns.color_palette()[2]
sns.countplot(data = flights, x = 'Airline', color = base_color)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(14,6))
base_color = sns.color_palette()[2]
sns.countplot(data = flights, y = 'Airline', color = base_color)
```

**Counting Missing Data:**

```python
flights.isna().sum()
```

```python
plt.figure(figsize=(14,6))
na_counts = flights.isna().sum()
base_color = sns.color_palette()[0]
sns.barplot(na_counts.index.values, na_counts, color=base_color )
plt.xticks(rotation=90)
plt.ylabel('Number of Missing values', fontsize = 12)
```

**Pie Charts:**

```python
plt.figure(figsize=(14,6))
sorted_counts = flights['Destination'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False)
plt.axis('square')
plt.title('Flight Destination')
```

```python
plt.figure(figsize=(14,6))
sorted_counts = flights['Destination'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, wedgeprops={'width': 0.4})
plt.axis('square')
plt.title('Flight Destination')
```

**Histograms:**

```python
plt.figure(figsize=(14,6))
plt.hist(data = flights, x = 'Duration(minutes)')
```

```python
plt.figure(figsize=(14,6))
plt.hist(data = flights, x = 'Price', bins = 20)
```

```python
plt.figure(figsize=(14,6))
bins = np.arange(0, flights['Price'].max()+1, 1200)
plt.hist(data = flights, x = 'Price', bins = bins)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.distplot(flights['Price'])
```

```python
plt.figure(figsize=(14,6))
sns.distplot(flights['Price'], kde = False)
```

```python
plt.figure(figsize=(14,6))
bin_edges = np.arange(0, flights['Price'].max()+1, 1200)
sns.distplot(flights['Price'], bins = bin_edges,  kde = False, hist_kws = {'alpha' : 1})
```
